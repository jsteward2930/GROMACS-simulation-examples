```python
mdp_content = """
; ions.mdp - used for adding ions to neutralize the system
integrator  = steep     ; Algorithm for energy minimization.
emtol       = 1000.0    ; Minimization stops when the maximum force < 1000.0 kJ/mol/nm
emstep      = 0.01      ; Energy minimization step size.
nsteps      = 50000     ; Maximum number of (minimization) steps.
nstlist     = 1         ; Frequency to update the neighbor list, every 1 steps
ns_type     = grid      ; Method to determine neighbor list (simple grid)
rlist       = 1.0       ; Cut-off for making neighbor list (short-range interactions)
coulombtype = cutoff    ; Simple cut-off scheme for Coulomb interactions
rcoulomb    = 1.0       ; Short-range electrostatic cut-off
rvdw        = 1.0       ; Short-range van der Waals cut-off
pbc         = xyz       ; Periodic boundary conditions in all directions
"""

with open('ions.mdp', 'w') as mdp_file:
    mdp_file.write(mdp_content)
```
